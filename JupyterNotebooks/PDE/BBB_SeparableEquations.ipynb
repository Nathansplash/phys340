# Separable Partial Differential Equations

Consider the following PDE:

$$
\frac{\partial^2 u}{\partial x^2} + \frac{\partial^2 u}{\partial y^2} = 0
$$

## Separation of Variables

We can separate the variables by assuming that $u(x,y) = X(x)Y(y)$, and then substituting this into the PDE:

$$
\frac{\partial^2 u}{\partial x^2} + \frac{\partial^2 u}{\partial y^2} = \frac{\partial^2 X}{\partial x^2}Y + X\frac{\partial^2 Y}{\partial y^2} = 0
$$

We can divide both sides by $XY$ to get:

$$
\frac{1}{X}\frac{\partial^2 X}{\partial x^2} + \frac{1}{Y}\frac{\partial^2 Y}{\partial y^2} = 0
$$


We can rearrange this to get:

$$
\frac{1}{X}\frac{\partial^2 X}{\partial x^2} = -\frac{1}{Y}\frac{\partial^2 Y}{\partial y^2} = \lambda
$$


Finally, we have now two separate ODEs:

$$
\frac{\partial^2 X}{\partial x^2} = \lambda X
$$

$$
\frac{\partial^2 Y}{\partial y^2} = -\lambda Y
$$

The first ODE has the solution:

$$
X(x) = A\cos(\sqrt{\lambda}x) + B\sin(\sqrt{\lambda}x)
$$

The second ODE has the solution:

$$
Y(y) = C\cosh(\sqrt{\lambda}y) + D\sinh(\sqrt{\lambda}y)
$$

where $cosh(x) = \frac{e^x + e^{-x}}{2}$ and $sinh(x) = \frac{e^x - e^{-x}}{2}$.  Of course, we could have also used the exponential form of the solution:

$$
Y(y) = C_1e^{\sqrt{\lambda}y} + C_2e^{-\sqrt{\lambda}y}
$$

Thus, the general solution to the PDE is:

$$
u(x,y) = (A\cos(\sqrt{\lambda}x) + B\sin(\sqrt{\lambda}x))(C\cosh(\sqrt{\lambda}y) + D\sinh(\sqrt{\lambda}y))
$$

We see that there are four arbitrary constants, and we can use the boundary conditions to determine these constants.

```python
# Example 1

import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 1, 100)
y = np.linspace(0, 1, 100)

X, Y = np.meshgrid(x, y)

A = 1
B = 1
C = 1
D = 1

lamb = 1

Z = (A*np.cos(np.sqrt(lamb)*X) + B*np.sin(np.sqrt(lamb)*X))*(C*np.cosh(np.sqrt(lamb)*Y) + D*np.sinh(np.sqrt(lamb)*Y))

plt.contourf(X, Y, Z)
plt.colorbar()
```
