# Laplace Equation

Laplace's Equation is a second order partial differential equation (PDE) that appears in many areas of science an engineering, such as electricity, fluid flow, and steady heat conduction. The equation is given by

$$\nabla^2 u = 0$$

where $\nabla^2$ is the Laplacian operator and $u$ is a function of position $x$ and $y$.

The Laplacian operator is defined as the divergence of the gradient of $u$:

$$\nabla^2 u = \nabla \cdot \nabla u = \frac{\partial^2 u}{\partial x^2} + \frac{\partial^2 u}{\partial y^2}$$

The Laplacian operator is a second order differential operator that maps a scalar function to a scalar function. The Laplacian is also the divergence of the gradient of a scalar function.

The Laplacian operator is used in many areas of science and engineering. For example, the Laplacian appears in Poisson's equation, which is used to describe electrostatics, and in the diffusion equation, which is used to describe heat conduction and fluid flow.

The Laplacian operator is also used in image processing, where it is used to compute the edges of an image. The Laplacian operator is also used in machine learning, where it is used to compute the Laplacian of the image and then used to compute the Laplacian of the image.

## Example: Laplace Equation in 2D

In this example, we will solve the Laplace equation in 2D using the finite difference method. The Laplace equation is given by

$$\nabla^2 u = 0$$

where $\nabla^2$ is the Laplacian operator and $u$ is a function of position $x$ and $y$.

This is a classic example of a PDE that can be solved using the finite difference meth.  It corresponds to a steady state heat conduction problem in 2D.  

The idea is to imagine a 2D square plate of material, with fixed temperature on the edges.  The temperature inside the plate will evolve until it reaches a steady state.  The temperature at any point inside the plate is given by the solution to the Laplace equation.

# Numerical Solution

The finite difference method is used to solve the Laplace equation numerically in 2D.  

We begin by defining the second derivative of $u$ with respect to $x$ and $y$:

$$\frac{\partial^2 u}{\partial x^2} = \frac{u(x+h,y) - 2u(x,y) + u(x-h,y)}{h^2}$$

$$\frac{\partial^2 u}{\partial y^2} = \frac{u(x,y+h) - 2u(x,y) + u(x,y-h)}{h^2}$$

where $h$ is the grid spacing.

We can then substitute these expressions into the Laplace equation to obtain:

$$\frac{u(x+h,y) - 2u(x,y) + u(x-h,y)}{h^2} + \frac{u(x,y+h) - 2u(x,y) + u(x,y-h)}{h^2} = 0$$

We can then rearrange this equation to obtain:

$$u(x,y) = \frac{1}{4} \left( u(x+h,y) + u(x-h,y) + u(x,y+h) + u(x,y-h) \right)$$

In other words, the value of $u$ at any point in the grid is equal to the average of the values of $u$ at the four neighboring points.

## Details

1. We will use a uniform grid with grid spacing $h$.
2. The grid will be a square grid with $N$ points in each direction.
3. We will specify the boundary conditions on the edges of the grid.
4. We will need to "guess" the value of $u$ at the interior points of the grid.

We then just iterate over the grid, computing the value of $u$ at each point using the equation above.  We continue iterating until the values of $u$ at the interior points converge to a steady state.



```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Set maximum iteration
maxIter = 500

# Set Dimension and delta
lenX = lenY = 20 #we set it rectangular
delta = 1

# Boundary condition
Ttop = 100
Tbottom = 0
Tleft = 0
Tright = 0

# Initial guess of interior grid
Tguess = 30
```

```python
# Set colour interpolation and colour map.
# You can try set it to 10, or 100 to see the difference
# You can also try: colourMap = plt.cm.coolwarm
colorinterpolation = 50
colourMap = plt.cm.jet

# Set meshgrid
X, Y = np.meshgrid(np.arange(0, lenX), np.arange(0, lenY))
```

```python
# Set array size and set the interior value with Tguess
T = np.empty((lenX, lenY))
T.fill(Tguess)

# Set Boundary condition
T[(lenY-1):, :] = Ttop
T[:1, :] = Tbottom
T[:, (lenX-1):] = Tright
T[:, :1] = Tleft
```

```python
# Iteration (We assume that the iteration is convergence in maxIter = 500)
print("Please wait for a moment")
for iteration in range(0, maxIter):
    for i in range(1, lenX-1, delta):
        for j in range(1, lenY-1, delta):
            T[i, j] = 0.25 * (T[i+1][j] + T[i-1][j] + T[i][j+1] + T[i][j-1])

print("Iteration finished")
```

```python
# Configure the contour
plt.title("Contour of Temperature")
plt.contourf(X, Y, T, colorinterpolation, cmap=colourMap)

# Set Colorbar
plt.colorbar()

# Show the result in the plot window
plt.show()

print("")
```
