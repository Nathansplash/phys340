# A Little Bit of Music Theory

This notebook is a brief introduction to some basic music theory concepts. It is intended to be a quick reference for people who are new to music theory and want to learn a little bit about it.

## Notes

In music, a note is a sound with a specific pitch. In Western music, there are 12 notes in an octave. These notes are named using the letters A, A#, B, C, C#, D, D#, E, F, F#, G, and G#.  The next note after G# is the A-note of the next octave. The notes repeat in the same order in each octave, and there are an infinite number of octaves.

In terms of frequency, an octave is a doubling of the frequency. For example, if the frequency of a note is 440 Hz, then the frequency of the same note in the next octave is 880 Hz, and the frequency of the same note in the previous octave is 220 Hz.

A good question is:  what are the frequencies of the notes in an octave?  The answer is that the frequencies of the notes in an octave are determined by the scale that is used. There are many different scales that are used in music, but the most common scale is the Western Even Tempered Scale.

## Western Even Tempered Scale

The Western Even Tempered Scale is a 12 note scale that is used in most Western music. It is called "even tempered" because the frequency ratio between each note is the same. This means that the frequency ratio between each note is the 12th root of 2. This is the same as saying that the frequency ratio between each note is 2^(1/12).

The frequency of the A-note in the so-called 4th octave is defined to be 440 Hz. This was a decision that was made in the 20th century by the International Organization for Standardization (ISO).  It was made to standardize the pitch of the A-note across the world.

This is called the "concert pitch". The frequency of the other notes in the scale are determined by the frequency of the A-note. The frequency of the A-note in the 5th octave is 880 Hz, and the frequency of the A-note in the 3rd octave is 220 Hz.

## Calculation of the Frequencies of the Notes in the Western Even Tempered Scale

The frequency of the A-note in the 4th octave is 440 Hz. The frequencies of the other notes above this A (i.e in the fifth octave) are calculated using the formula:

f = 440 * $2^{n/12}$

where f is the frequency of the note, and n is the number of half steps between the note and the A-note. The following table shows the frequencies of the notes in the 4th/5th octaves (note, for good reasons, the octaves are grouped together starting with C, and not A):

| Note | Frequency (Hz) |  Ratio |
|------|----------------|--------|
| A4   | 440            | 1.0    |
| A#4  | 466.16         | 1.059  |
| B4   | 493.88         | 1.122  |
| C5   | 523.25         | 1.189  |
| C#5  | 554.37         | 1.259  |
| D5   | 587.33         | 1.334  |
| D#5  | 622.25         | 1.414  |

These frequencies are the frequencies of the notes in the 4th octave. We can thus calculate the frequency of any note in any octave using these values, together with appropriate factors of 1/2 or 2.

For example, a so-called "high C" is the C-note in the 6th octave.  The frequency of the high C is 523.25 * 2 = 1046.5 Hz.  This is an interesting note because it has, historically, been used as a measure of the singing ability of sopranos.  The ability to sing the high C is considered to be a sign of a very good soprano.

For bass singers, the low E is often used as a measure of singing ability.  The low E is the E-note in the 2nd octave.  The frequency of the low E is 82.41 Hz (i.e. 659.25 Hz * 1/2 * 1/2 * 1/2 * 1/2)

# Why are there 12 notes in an octave?

It is really an interesting question, is it not?  I mean, why are there 12 notes in an octave?  Why not 10, or 20, or 100?  

Part of the  answer to this question is that the number 12 is special because it is has a large number of divisors.  The divisors of 12 are 1, 2, 3, 4, 6, and 12.  We call numbers that have a larger number of divisors than any number smaller than that number $highly complex numbers$.  The number 12 is a highly complex number.

The Greeks realized that sounds which have frequencies in rational proportion are perceived as harmonius. For example, if the frequency of one note is 440 Hz, and the frequency of another note is 660 Hz, then the ratio of the frequencies is 3:2.  This is a rational number.  The Greeks realized that the most harmonius sounds are those that have frequencies in rational proportion.  They also realized that the most harmonius sounds are those that have frequencies in the $simplest$ rational proportion.  The simplest rational proportion, other than unity, is 2:1.  This is the ratio of the frequencies of the notes in an octave!

Let's look at some of the simplest ratios of the frequencies of the notes in an octave:

Notice that the ratios of the frequencies of the notes in an octave (in the even tempered scale) are all close to the simplest rational numbers.  This is why the even tempered scale is so harmonius. 

But, obviously there are other choices that could have been made.  For example, the Greeks used a scale called the Pythagorean scale.  In the Pythagorean scale, the ratio of the frequencies of the notes in an octave are all powers of 3/2.  This is a very harmonius scale, but it has some problems.  For example, the Pythagorean scale has a note called the "wolf note".  This note is so called because it sounds terrible.  The wolf note is the note that is 23.46 cents higher than the note that is 7 octaves below it.  This is a problem because it means that the Pythagorean scale is not a "closed" scale.  This means that it is not possible to play the scale in all keys.  The even tempered scale is a closed scale, and this is one of the reasons why it is so popular.


## Single Ratio

The ratio of the frequencies of the notes in an octave is 2:1.  The next simplest ratio is 3:2.  This is the ratio of the frequencies of the notes that are a perfect fifth apart.  The perfect fifth is a very harmonius interval.  It is the interval that is used to tune the strings of a guitar, the strings of a violin, and the strings of a piano.  The perfect fifth is a very important interval in music.  And, it is the key concept in understanding why there are 12 notes in an octave.

## The Circle of Fifths

Looking at the order of the notes in the table above, starting from C, we then get G D A E B F# C# Ab Eb Bb F and back to C).
The chromatic scale reflects this fact. In the 18th and 19th centuries, the chromatic scale was tuned using the idea of 3/2. In the most elegant of these, Thomas Young's tuning, several of the fifths were set exactly to 3/2, and the others were tempered slightly (to make octaves exact).

The circle of fifths is a diagram that shows the relationships between the notes in the even tempered scale.  It is called the circle of fifths because it shows the relationships between the notes in the scale that are a perfect fifth apart.  The circle of fifths is a very useful tool for understanding the relationships between the notes in the scale, and for understanding the relationships between the keys in the scale.

## Putting it all together

Okay, so we come the final step in understanding "Why 12 Notes in and Octave".

It comes down, finally, to the idea that we would like, in principle, to have the ratio 3:2, raised to some power, be EXACTLY equal to 2.  

This is not possible, but we can get close. 

$$
\left(\frac{3}{2}\right)^a = 2^b
$$

where a and b are integers.  This is the same as saying that:

$$
\frac{3}{2} = 2^{\frac{b}{a}} = 2^{x}
$$

where x is a rational number.  Solving for x, we get:

$$
x = \frac{\log (\frac{3}{2})}{\log 2} = 0.5849625007211562
$$

So, the questions is, what RATIONAL value of x is closest to this value?  

The answer is 7/12 = 0.583333.  

This is a value of x that is close to the value of x that we calculated above, and also has a denominator of 12, which is a highly composite number.  This is the reason why there are 12 notes in an octave.  It is because the ratio of the frequencies of the notes in an octave is close to the simplest rational number that is close to the ratio of the frequencies of the notes that are a perfect fifth apart.
